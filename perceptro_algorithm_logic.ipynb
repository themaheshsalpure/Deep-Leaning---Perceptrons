```python
from sklearn.datasets import make_classification
import numpy as np

X, y = make_classification(n_samples=100, n_classes=2, n_features=2, n_informative=1, n_redundant=0, n_clusters_per_class=1, random_state=41, hypercube = False, class_sep = 10)
```

```python
X
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (8, 5))
sns.scatterplot(x = X[:,0],y =  X[:,1],hue=y)
```

```python
def perceptron(X, y):

    # here we inserted the column completely having values only one for X0 
    # we inserted in X at position 0 values inserted will be 1 in axis = columns
    X = np.insert(X, 0, 1, axis = 1)

    # at initial we will select the random weights for plotting an random line in the plot.
    weights = np.ones(X.shape[1])
    lr = 0.1

    for i in range(200):
        j = np.random.randint(100)
        y_hat = step(np.dot(X[j],weights))
        weights= weights + lr*(y[j]-y_hat)*X[j]

    return weights[0], weights[1:]
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_, coef_ = perceptron(X, y)
```

```python
intercept_
```

```python
print(coef_)
```

Finding the m and b for y = mx + b  
here, m = -(A/B)   
and b = -(C/B)  

Here, these A and B are 2 values of the coef_ and the remaining C is values of the intercept_

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize = (10, 6))
plt.plot(x_input, y_input, color = 'red', linewidth = 2)
plt.scatter(X[:,0], X[:,1], c= y, cmap='winter')
plt.ylim(-3, 2)
```
